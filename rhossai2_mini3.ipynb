<h1>Correlation Analysis</h1>

To find the relationships among data columns we need to analyze the the correlation. 

```r
sel = c()
for (i in 1:dim(data)[2]) if (is.numeric(data[,i])) sel = c(sel, i);

d =data[,sel]

cor(d,method="spearman",use="pairwise.complete.obs");
```

```r
high_corr <- function(x, level){
  res <- cor(x, method="spearman");
  res1 <- res; res1[res<0] <- -res[res < 0];
  for (i in 1:dim(x)[2]){
    res1[i,i] <- 0;
  }
  sel <- apply(res1,1,max) > level;
  res[sel,sel];
}
high_corr(d,.7)
```

<h1>Normalizing and removing highly corellated data</h1>

There are some variables which seem to be overly correlated. To get a better approximation these variables need to be removed from the dataset.

<h1>  Fitting the Statistical Model- Regression Model</h1>



```r
res <- c();
vnam <- names(d);
for (i in 2:dim(d)[2]){
  fmla <- as.formula(paste(vnam[i],paste(vnam[-c(1,i)],collapse="+"),sep="~"));
  res <- rbind(res,c(i,round(summary(lm(fmla,data=d))$r.squared,2)));
}
row.names(res) <- vnam[res[,1]];
res[order(-res[,2]),]
```
