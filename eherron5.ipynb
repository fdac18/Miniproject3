A Survey on Technology Choice
======


    For this assignment, a variety of models were created and tested for the survey response variable of PG5_7NDYP. This variable represents the ranked importance of the number of developers in each survey response. Prior to begining the analysis it was hypothesized that the varibles most predictive of this response variable would be PG0Shown, the number of developers shown; PG1PsnUse, project for personal versus research use; and PG5_6SSYOP, the scale and size of the project. After numerically encoding all factor type attributes and normalizing the dataset, the cart algorithm was used to predict the response variable using the three variables selected in the hypothesis, yielding a test set accuracy of 0.80826. 
    The actual best attributes for predicting the response variable were determined using two approaches: dropping one attribute at a time from the model and using linear regression. From the first approach, it was determined that the accuracies of the models were lowest when PG3Submit, PG5_11NSG, PG0Shown, PG0Submit, and PG5_7Time were dropped, indicating the these were most useful attributes for predicting the response variable. For the second approach, PG5_7Order, PG5_2BNUI, PG5_12Order, PG5_13Order, PG5_11Order, PG5_4VGP, PG5_1RRPQ, PG5_8CP, PG5_6SSYOP, and PG5_2Time were found to have the strongest correlation to PG5_7NDYP. The groups of attributes selected from each approach were used to train a CART model. The test accuracy observed for the attributes from the first method was 0.87316, while the test accuracy obtained from the second approach was 0.85546.
    These results differ from the hypothesized outcome and indicate that the elimination approach was best for determining the attributes most predictive of the response variable. Of the attributes predicted to be most useful for modelling PG5_7NDYP, only PG0Shown was determined to have been significantly useful. The other attributes selected from the results of the elimination method were PG3Submit, the submission time of page 3; PG5_11NSG, the rank of the importance of the number of stars on Github; PG0Submit, the submission time of page 0, and PG5_7Time, the completion time of factor 7 on page 5. 
    In order to improve the survey in the future, the ranking question on page 5 could be changed to a page containing a series of multiple choice questions asking survey takers to rate each individual factor by its importance. This change would result in clearer data for those analyzing the results from this page.

```r
# Read in the data
data <- read.csv("TechSurvey - Survey.csv",header=T);

#convert date to unix second
for (i in c("Start", "End")) 
    data[,i] = as.numeric(as.POSIXct(strptime(data[,i], "%Y-%m-%d %H:%M:%S")))
for (i in 0:12){
    vnam = paste(c("PG",i,"Submit"), collapse="")
    data[,vnam] = as.numeric(as.POSIXct(strptime(data[,vnam], "%Y-%m-%d %H:%M:%S")))
}
#calculate differences in time    
for (i in 12:0){
    pv = paste(c("PG",i-1,"Submit"), collapse="");
    if (i==0) 
        pv="Start";
    vnam = paste(c("PG",i,"Submit"), collapse="");
    data[,vnam] = data[,vnam] -data[,pv];
}
```

### Interpret  basic summaries

```r
#get numeric fields only for correlation
sel = c()
for (i in 1:dim(data)[2]) if (is.numeric(data[,i])) sel = c(sel, i);
num_data = data[,sel]
```

```r
cor(num_data,method="spearman",use="pairwise.complete.obs"); #OK for any: uses ranks
```

```r
#just see top corelations
hiCor <- function(x, level){
  res <- cor(x,method="spearman", use='pairwise.complete.obs');
  res1 <- res; res1[res<0] <- -res[res < 0];
  for (i in 1:dim(x)[2]){
    res1[i,i] <- 0;
  }
  sel <- apply(res1,1,max) > level;
  res[sel,sel];
}
hiCor(num_data,.7)
```

```r
plot(1:dim(num_data)[2],cumsum(prcomp(num_data, retx=F,scale=T)$sdev^2)/sum(prcomp(num_data, retx=F,scale=T)$sdev^2),ylim=c(0,1),xlab="Number of coponents",ylab="Fraction of variance");
res<-prcomp(data[,sel], retx=F,scale=T)$rotation[,1:5];
resAbs <- res;
resAbs[res<0] <- -res[res<0];
for (i in 1:5)
  print(t(res[resAbs[,i]>.3,i,drop=FALSE]));
```

```r
#regress each predictor on the remaining predictors
# eliminate with the highest adjR^2
res <- c();
vnam <- names(num_data);
for (i in 2:dim(num_data)[2]){
  #print(paste(vnam[i],paste(vnam[-c(1,i)],collapse="+"),sep="~"));
  fmla <- as.formula(paste(vnam[i],paste(vnam[-c(1,i)],collapse="+"),sep="~"));
  res <- rbind(res,c(i,round(summary(lm(fmla,data=num_data))$r.squared,2)));
}
row.names(res) <- vnam[res[,1]];
res[order(-res[,2]),];
```
